1. Отримати всі профіля з назвою професії

```python
sql = '''
select name,	lastname,	profession.profession_name
from profile
join profession on profile.profession_id = profession.id

'''

get_select(sql)
```

2. Написати запит який відобразить назви всіх вакансії кожної з компаній

```python
sql = '''
select vacancy_name, employer.employer_name
from vacancy
join employer on vacancy.employer_id = employer.id

'''

get_select(sql)
```

3. Визначити кількість профілів по кожній професії


```python
sql = '''
select profession_name, count(*) as profile_count
from profile
join profession on profile.profession_id = profession.id
group by profession_name


'''

get_select(sql)
```

4. Визначити кількість вакансій опублікованих кожною компанією

```python
sql = '''
select employer_name, count(vacancy_name) as vacancy_count
from vacancy
join employer on vacancy.employer_id = employer.id
group by employer_name


'''

get_select(sql)
```

5. Визначити компанію, яка не опублікувала жодної вакансії

```python
sql = '''
select employer_name, vacancy_name
from employer
left join vacancy on vacancy.employer_id = employer.id
where vacancy.id is null


'''

get_select(sql)
```

6. Відобразити всі унікальні вакансії (бажано 2-ма способами)


```python
sql = '''
select vacancy_name, count(*)
from vacancy
group by vacancy_name


-- select distinct vacancy_name from vacancy

'''

get_select(sql)
```

7. Вивести з таблиці профілів максимальний id та дату його створення

```python
sql = '''
select name, lastname, create_date, max(id) as max_id
from profile

'''

get_select(sql)
```

8. Вивести всі компанії в яких більше 3 розміщених вакансій


```python
sql = '''
select employer_name , count(vacancy_name) as count_vacancy
from employer
join vacancy on vacancy.employer_id = employer.id
group by employer_name
having count_vacancy > 3
'''

get_select(sql)
```

1. конверсію цієї функції (кількість додавань в закладки по відношенню до кількості переглядів сторінки
товару)


```python
sql = '''
select
    ppv.id_product,
    type,
    count(ppa.id) as count_save,  -- кількість додавань в закладки
    count(ppv.id) as count_page_views, -- кількість переглядів сторінки товару
    (count(ppa.id) / count(ppv.id)) as conversion
from
    product_page_view as ppv
left join
    product_page_action as ppa on ppv.id = ppa.id_page_view
where ppa.type = 2
group by
    ppv.id_product

'''

get_select(sql)
```

2. ту саму конверсію, але тільки для тих показів сторінки товару, коли користувач зареєстрований і у нас є
інформація про те що він/вона раніше виконали хоча б одну купівлю

```python
sql = '''
select
    ppv.id_product,
    count(ppa.id) as count_save,  -- кількість додавань в закладки
    count(ppv.id) as count_page_views, -- кількість переглядів сторінки товару
    (count(ppa.id) / count(ppv.id)) as conversion
from
    product_page_view as ppv
join
    product_page_action ppa on ppv.id = ppa.id_page_view
join
    transaction_data as td on ppa.id_page_view = td.id_user_account

where ppa.type = 2 and td.id_user_account not null
group by
    ppv.id_product


'''

get_select(sql)
```
